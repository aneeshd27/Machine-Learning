```python
data['Score']=pd.to_numeric(data['Score'],errors='coerce')
data=data.dropna(subset=['Score'])
data.head()
```

```python
sns.histplot(x=data['Score'],color='violet')
plt.title('Century Scores')
```

```python
sns.scatterplot(x=data['Position'],y=data['Score'])
plt.title('Scatterplot for Score VS Position')
```

```python
sns.scatterplot(x=data['Position'],y=data['Score'],hue=data['Result'])#size=25.0
plt.title('Scatterplot for Score VS Position')
```

```python
data['Date']=pd.to_datetime(data['Date'],dayfirst=True)
plt.figure(figsize=(10,8))
data.sort_values(['Date'],inplace=True)
plt.plot(data['Date'],data['Score'],marker='o')
plt.title("Timeline of Steve Smith's Centuries")
plt.xticks(rotation=45)
```

```python
result=data.groupby(['Against','Result']).size().unstack()
result.plot(kind='bar',stacked=True,figsize=(12,6))
plt.xlabel('Opponent')
plt.ylabel('Number of Centuries')
plt.title("Steve Smith's Centuries against Diff Opponents and their results")
plt.show()
```

```python
runs=data.groupby('Venue')['Score'].sum()
top10runs_grounds=runs.sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
plt.bar(top10runs_grounds.index,top10runs_grounds.values)
plt.xlabel('Grounds')
plt.ylabel('Total Runs')
plt.title('Top 10 Venues where Steve Smith likes to play')
plt.xticks(rotation=90)
```

```python
data['Total Score']=data['Score'].cumsum()
plt.figure(figsize=(12,6))
plt.plot(data['Date'],data['Total Score'],marker='o')
plt.xlabel('Year')
plt.ylabel('Consisitency Over the years')
plt.xticks(rotation=35)
plt.show()
```
